```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(-3, 3, 0.1)
y1 = 10*x**3 - x
y2 = 10**3*np.cos(x)**2

plt.plot(x, y1, 'b-')

plt.figure()
plt.plot(x, y2, 'r-.')

plt.show()
```

```python
#plt.plot(x, y1, 'b-')
#plt.plot(x, y2, 'r-.')
plt.plot(x, y1, 'b-', x, y2, 'r-.')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid(True) # plt.grid()
plt.legend(['y1', 'y2'])

plt.show()
```
